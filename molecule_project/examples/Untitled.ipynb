```python
from molecules.molecule import Molecule
from molecules.geometry import distance

# Load water, methane, ethanol molecules from XYZ files
water = Molecule().from_xyz("water.xyz")
methane = Molecule().from_xyz("methane.xyz")
ethanol = Molecule().from_xyz("ethanol.xyz")

# Inspect water molecule
print("Initial water molecule:")
print(water)
```

```python
# Translate water by vector (1.0, 0.0, 0.0)
water.translate((1.0, 0.0, 0.0))

print("Water molecule after translation by (1.0, 0.0, 0.0):")
print(water)
```

```python
# Rotate water around z-axis by 90 degrees
water.rotate(origin=(0,0,0), axis=(0,0,1), angle_deg=90)

print("Water molecule after rotation 90° around z-axis:")
print(water)
```

```python
water.stretch_bond(0, 1, 0.1)
water
```

```python
a, b = water.atoms[0], water.atoms[1]
dist = distance(a.coords, b.coords)
dist
```
